<h5><font color = 'yellow'> Step 1 – data collection</font></h5>
<h5>We are scraping data from the Wikipedia site</h5>

```python
url = 'https://en.wikipedia.org/wiki/Template:COVID-19_data#cite_note-3'
req = requests.get(url)                                     # get method of requests library and this will fetch the data from url
```

```python
data_list = pd.read_html(req.text)                          # to filter out the data into DF we are using read_html fuction
```

```python
target_df = data_list[0]
target_df
```

```python
# lets check top records
target_df.head()
```

```python
# lets check bottom records
target_df.tail()
```

<h5> <font color = 'Yellow'>Step 2 - Data Cleaning </font> </h5>
<h6> let's figure out the problems--</h6>
<h6>1)	delete column name "Unnamed:0"</h6>
<h6>2)  delete 1st 2rows</h6>
<h6>3)	delete last row</h6>
<h6>4)	replace “–“ from all Columns to 0</h6>
<h6>5)	convert the string data type to number wherever the numbers are showing</h6>

```python
# lets change the column name
target_df = target_df.rename(columns={'Vaccinated[a]':'Vaccinated', 'Vaccine doses[b]':'Vaccine doses',
                                        'Fully vaccinated[c]':'Fully vaccinated'})
target_df
```

```python
# lets solve the 1st problem 
# remove the column -"Unnamed-0"

target_df = target_df[['Location', 'Cases','Deaths', 'Vaccinated','Vaccine doses','Fully vaccinated']]
target_df.head()
```

```python
target_df.tail()
```

```python
# # lets remove the first 2 rows & remove last row

last_idx = target_df.index[-1]          # this method is used to delete the last row dynamically
target_df = target_df.drop([0,1,last_idx])
target_df
```

```python
target_df.reset_index(drop=True, inplace=True)
target_df
```

```python
# lets replace "—" from all columns

target_df['Cases'] = target_df['Cases'].str.replace('—','0')
target_df['Deaths'] = target_df['Deaths'].str.replace('—','0')
target_df['Vaccinated'] = target_df['Vaccinated'].str.replace('—','0')
target_df['Vaccine doses'] = target_df['Vaccine doses'].str.replace('—','0')
target_df['Fully vaccinated'] = target_df['Vaccine doses'].str.replace('—','0')

target_df
```

```python
target_df.dtypes
```

```python
# lets fixed the data type of all those values and convert to numeric

target_df['Cases'] = pd.to_numeric(target_df['Cases'])
target_df['Deaths'] = pd.to_numeric(target_df['Deaths'])
target_df['Vaccinated'] = pd.to_numeric(target_df['Vaccinated'])
target_df['Vaccine doses'] = pd.to_numeric(target_df['Vaccine doses'])
target_df['Fully vaccinated'] = pd.to_numeric(target_df['Fully vaccinated'])
```
